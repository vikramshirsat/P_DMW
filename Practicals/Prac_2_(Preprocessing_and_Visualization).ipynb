```python
# Data Cleaning
#1. Finding Duplicated Values
df.duplicated()
```

```python
#2. Filling Null Values
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```
