```python
#Preprocessing.
df.isnull().sum()
```

```python
#Filling Missing Valus of Age Column
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```
